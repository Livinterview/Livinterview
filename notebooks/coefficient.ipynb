```python
import pandas as pd

final_df = pd.read_csv("data/final.csv", encoding="utf-8")
final_df.drop("인구대비 갯수", axis=1, inplace=True)
```

```python
final_df.rename(columns={"비율": "계수"}, inplace=True)
final_df.head()
```

```python
per_count = [
    "약국",
    "따릉이대여소",
    "버스정류장",
    "지하철역",
    "공원",
    "공연장",
    "문화공간",
    "백화점",
    "시장",
]
per_population = [
    "셀프빨래방",
    "영화관",
    "코인노래방",
    "PC방",
    "공공체육시설",
    "헬스장",
    "양방",
    "한방",
    "카페",
    "경찰",
    "도서관",
    "은행",
    "주민센터",
]
per_area = [
    "편의점",
    "대형마트",
    "다이소",
    "반찬가게",
    "소방",
]
```

```python
pop_df = pd.read_csv("data/population.csv", encoding="utf-8")
pop_df = pop_df[pop_df["시도명"] == "서울특별시"]
pop_df = pop_df[["시군구명", "읍면동명", "계"]]
# pop_df
print(len(pop_df["읍면동명"]), *pop_df["읍면동명"])
print(len(final_df["동"].unique()), *final_df["동"].unique())
```

```python
# print(len(pop_df["읍면동명"]), *pop_df["읍면동명"])
# print(len(final_df["동"].unique()), *final_df["동"].unique())

subtract_set = set(final_df["동"].unique()) - set(pop_df["읍면동명"])
subtract_set
```

```python
print("원래 동 개수:", len(final_df["동"].unique()))
print("제외할 동 개수:", len(subtract_set))

filtered_df = final_df[~final_df["동"].isin(subtract_set)]
print("제거 후 동 개수:", len(filtered_df["동"].unique()))

print("실제 제거된 동 이름들:", subtract_set)
```

```python
for category in per_count:
    mask = filtered_df["소분류"] == category
    filtered_df.loc[mask, "계수"] = filtered_df.loc[mask, "개수"]
```

```python
pop_df = pd.read_csv("data/population.csv", encoding="utf-8")
pop_df = pop_df[pop_df["시도명"] == "서울특별시"]
pop_df = pop_df[["읍면동명", "계"]]
pop_df.head()
```

```python
dong_to_population = dict(zip(pop_df["읍면동명"], pop_df["계"]))
dong_to_population
```

```python
pop_df = pd.read_csv("data/population.csv", encoding="utf-8")
pop_df = pop_df[pop_df["시도명"] == "서울특별시"]
pop_df = pop_df[["시군구명", "읍면동명", "계"]]
pop_df
for category in per_population:
    mask = filtered_df["소분류"] == category
    filtered_df.loc[mask, "계수"] = filtered_df.loc[mask, "개수"] / filtered_df.loc[
        mask, "동"
    ].map(dong_to_population)
filtered_df[filtered_df["소분류"] == "영화관"]
```

```python
area_df = pd.read_csv("data/area.csv", encoding="utf-8")
dong_to_area = dict(zip(area_df["EMD_NM"], area_df["면적_km2"]))
dong_to_area
```

```python
for category in per_area:
    mask = filtered_df["소분류"] == category
    filtered_df.loc[mask, "계수"] = filtered_df.loc[mask, "개수"] / filtered_df.loc[
        mask, "동"
    ].map(dong_to_area)
filtered_df[filtered_df["소분류"] == "편의점"]
```

```python
filtered_df.to_csv("data/final.csv", encoding="utf-8", index=False)
```
