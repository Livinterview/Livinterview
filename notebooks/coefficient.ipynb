```python
per_count = [
    "약국",
    "따릉이대여소",
    "버스정류장",
    "지하철역",
    "공원",
    "공연장",
    "문화공간",
    "백화점",
    "시장",
]
per_population = [
    "셀프빨래방",
    "영화관",
    "코인노래방",
    "PC방",
    "공공체육시설",
    "헬스장",
    "양방",
    "한방",
    "카페",
    "경찰",
    "도서관",
    "은행",
    "주민센터",
]
per_area = [
    "편의점",
    "대형마트",
    "다이소",
    "반찬가게",
    "소방",
]
```

```python
pop_df = pd.read_csv("data/population.csv", encoding="utf-8")
pop_df = pop_df[pop_df["시도명"] == "서울특별시"]
pop_df = pop_df[["시군구명", "읍면동명", "계"]]
pop_df
```

```python
dong_to_population = dict(
    zip(pop_df[["시군구명", "읍면동명"]].apply(tuple, axis=1), pop_df["계"])
)
dong_to_population
```

```python
filtered_df.loc[filtered_df["소분류"].isin(per_population), "계수"] = pd.NA
```

```python
filtered_df[filtered_df["소분류"].isin(per_population)]
```

```python
for category in per_population:
    mask = filtered_df["소분류"] == category

    # map으로 인구수를 가져오되, 인덱스를 유지하도록 Series로 처리
    keys = list(
        zip(
            filtered_df.loc[mask, "구"].str.strip(),
            filtered_df.loc[mask, "동"].str.strip(),
        )
    )

    populations = pd.Series(
        [dong_to_population.get(k) for k in keys],
        index=filtered_df.loc[mask].index,  # 인덱스를 유지!
    )

    filtered_df.loc[mask, "계수"] = filtered_df.loc[mask, "개수"] / populations


filtered_df[filtered_df["소분류"] == "영화관"]
```

```python
area_df = pd.read_csv("data/area.csv", encoding="utf-8")
dong_to_area = dict(
    zip(area_df[["자치구", "EMD_NM"]].apply(tuple, axis=1), area_df["면적_km2"])
)
dong_to_area
```

```python
for category in per_area:
    mask = filtered_df["소분류"] == category

    keys = list(
        zip(
            filtered_df.loc[mask, "구"].str.strip(),
            filtered_df.loc[mask, "동"].str.strip(),
        )
    )

    areas = pd.Series(
        [dong_to_area.get(k) for k in keys],
        index=filtered_df.loc[mask].index,  # 인덱스를 유지!
    )

    filtered_df.loc[mask, "계수"] = filtered_df.loc[mask, "개수"] / areas

filtered_df[filtered_df["소분류"] == "편의점"]
```
